```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

csv = pd.read_csv("https://raw.githubusercontent.com/NumericalMethodsSullivan/NumericalMethodsSullivan.github.io/master/data/Exercise3_datafit4.csv", header = None)

# Define a function to compute the first derivative using finite differences
def FirstDeriv(a,b):
    derList = []
    x_list = []
    x= a
    h = b
    # Iterate through the list of values, except the last one, to compute finite difference
    for i in range(len(h)-1):
      # Compute the difference quotient (approximate derivative)
      df = (h[i+1]-h[i])/(x[i+1]-x[i])
      x_list.append(x[i+1])
      derList.append(df)
    # Return the list of x values and the corresponding derivative values
    return x_list, derList

x = csv.iloc[:,0]
y = csv.iloc[:,1]
# Plot the original data points (blue line)
plt.plot(x, y, 'b')

# Compute the first derivative of the data
x_a, y_a = FirstDeriv(x, y)

# Plot the approximated first derivative (red line)
plt.plot(x_a, y_a, 'r-')

plt.grid()
plt.legend(['exact first deriv',
            'approx first deriv'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

csv = pd.read_csv("https://raw.githubusercontent.com/NumericalMethodsSullivan/NumericalMethodsSullivan.github.io/master/data/Exercise3_datafit4.csv", header = None)

# Define a function to compute the second derivative using finite differences
def SecondDeriv(a,b):
    derList = []
    x_list = []
    x= a
    h = b
    for i in range(1, len(h)-1):
      # Compute the second difference quotient (approximate second derivative)
      # Formula: (y(i+1) - 2*y(i) + y(i-1)) / (x(i+1) - x(i))^2
      df = (h[i+1] - (2*h[i]) + h[i-1])/((x[i+1]-x[i])**2)
      x_list.append(x[i+1])
      derList.append(df)
    return x_list, derList

x = csv.iloc[:,0]
y = csv.iloc[:,1]
plt.plot(x,y, 'b')

x_sdf, y_sdf = SecondDeriv(x,y)
plt.plot (x_sdf, y_sdf, 'g-',)
plt.grid()
plt.legend(['f(x) = sin(x)',
            'approx second deriv'])
plt.show()
```
