Finding the integral using the trapezoid rule

```python
import numpy as np
import matplotlib.pyplot as plt


#same function as above, but different methods used for the Trapeziod

def riemannSum(f:'function', lower: int, upper:int, interval: int, riemann: str ='middle'):
  points = np.linspace(lower, upper, interval+1)
  n = points[1] - points[0]
  list_riemann = []
  if riemann == 'middle':
    half = n/2
    #right side -left side function * interval
    for i in range(len(points)-1):
      mid = points[i] + half
      list_riemann.append(f(mid)*(n))
  elif riemann == 'left':
    #right side -left side function * interval
    for i in range(len(points)-1):
      left = points[i]
      list_riemann.append(f(left)*(n))
  elif riemann == 'right':
    #right side -left side function * interval
    for i in range(1,len(points)):
      right = points[i]
      list_riemann.append(f(right)*(n))

  elif riemann == 'trapezoid':
    for i in range(len(points)-1):
      trap = .5*(f(points[i+1])+f(points[i]))*(points[i+1]-points[i])
      list_riemann.append(trap)

    return sum(list_riemann)
  else:
    raise TypeError(f"You can either make riemann equal 'middle', 'left', 'right', 'trapezoid_google', or 'trapezoid_book'")

  return sum(list_riemann)
```

```python
#Test functions

f = lambda x: x/(1+x**4)

i = lambda x: (x-1)**3 * (x-2)**2

j = lambda x: np.sin(x**2)


approx1 = riemannSum(f,-1,2,100000,'trapezoid')
#The exact value is :0.27020
approx2 = riemannSum(i,-1,2,100000,'trapezoid')
#The exact value is -27.45
approx3 = riemannSum(j,-1,2,100000,'trapezoid')
#The exact value is 1.11504

print('#'*30)
print(f'Trapezoid_f = {approx1}\nTrapezoid_i= {approx2}\nTrapezoid_j = {approx3}')
print('#'*30)

```

```python
#First derivative

import numpy as np
import matplotlib.pyplot as plt
def FirstDeriv(f,a,b,N):
    x = np.linspace(a,b,N+1) # Partition the interval [a,b] into N subintervals
    h = x[1] - x[0] # This finds h = Delta x
    df = (f(x[1:])-f(x[:-1]))/h
    return df

f = lambda x: x/(1+x**4)
exact_df = lambda x: (1-3*x**4)/(1+x**4)**2
a = -1
b = 2
N = 100
x = np.linspace(a,b,N+1)
df = FirstDeriv(f,a,b,N)
plt.figure(figsize=[12,8])
plt.plot(x,f(x),'b',x,exact_df(x),'r--',x[:-1],df,'k-.')
plt.grid()
plt.legend(['f(x) = x/(1+x**4)','exact first deriv','approx first deriv'])
plt.show()
```
