1. Create a basin on attraction image for the function $f(x) = (x-4)(x+1)$

```python
#Question 1
import numpy as np
import matplotlib.pyplot as plt

def newton(f,df,x0,tol=1e-8):
  x = x0
  counter = 0
  while np.abs(f(x))>tol and counter < 1000:
      if (df(x0) == 0):
        print("Pick a different x0")
        return False
      else:
        x = (-f(x) + df(x)*(x))/df(x)
        x = np.round(x)
        if x == -1:
          plt.plot(x0,0, 'g', marker=".", markersize=2)
        if x == 4:
          plt.plot(x0,0, 'r', marker=".", markersize=2)
        counter += 1 # check for runaway loops
        print("The root using Newton's method is", x)



r = np.linspace(-5, 8, 75)
for i in range(len(r)):
  f = lambda x: (x-4)*(x+1)
  df = lambda x: 2*x - 3
  newton(f,df, (r[i]))

plt.plot(r, (r-4)*(r+1), 'b--')
plt.plot(-1,0, 'go')
plt.plot(4,0, 'ro')
plt.grid()
plt.show()
```

5. Your job is to build a picture of the basins of attraction for the three roots in the complex plane. When you have your picture give a thorough write up of what you found.

We find our graph is divided into three equal size sections where each section converges to one of our roots. The three "lines", or rather clusters of points, that separate the major sections all have a similar pattern to them, but each separator arranges the colors of the pattern differently. All the separator "lines" are surrounded by a thin layer of orange on either side. Ignoring these thin layers of orange, we see a clockwise pattern of orange to green to blue and back to orange in both the major chunks of the image and in the larger chunks of the separators.

```python
#question 5

def newton(f,df,x0,tol=1e-8):
  x = x0
  counter = 0
  while np.abs(f(x))>tol and counter < 1000:
      if (df(x0) == 0):
        break
      else:
        x = (-f(x) + df(x)*(x))/df(x)
        counter +=1
  return x




f = lambda x: x**3-1
df = lambda x: 3*(x**2)
roots = [[1, 0, 'red'], [-0.5, 0.5*np.sqrt(3), 'green'], [-0.5, -0.5*np.sqrt(3), 'blue']]


x_r = np.linspace(-2, 2, 400)
x_i = x_r*1j
#grid_r, grid_i = np.meshgrid(z_r, z_i)
#z = np.multiply(grid_i, grid_r)
emptyS=[]
for i in range(len(x_r)):
  for k in range(len(x_i)):
    emptyS.append(x_i[k]+x_r[i])

blue = [[],[]]
green = [[],[]]
orange = [[],[]]

for i in range(len(emptyS)):
  root = newton(f,df, emptyS[i])
  if np.round(root, 4) == np.round(roots[0][0]+roots[0][1]*1j, 4):
    blue[0].append(np.real(emptyS[i]))
    blue[1].append(np.imag(emptyS[i]))
  elif np.round(root, 4) == np.round(roots[1][0]+roots[1][1]*1j, 4):
    green[0].append(np.real(emptyS[i]))
    green[1].append(np.imag(emptyS[i]))
#  elif np.round(root, 4) == np.round(roots[2][0]+roots[2][1]*1j, 4):
  else:
    orange[0].append(np.real(emptyS[i]))
    orange[1].append(np.imag(emptyS[i]))

#plt.plot(x, f(x), 'k--')

plt.axis('off')
plt.scatter(blue[0], blue[1], color = 'blue', s = 1)
plt.scatter(green[0], green[1], color = 'green', s = 1)
plt.scatter(orange[0], orange[1], color = 'orange', s= 1)
```
